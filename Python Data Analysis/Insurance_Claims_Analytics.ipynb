## 11.1 Univariate Analysis

Univariate Analysis examines one variable at a time to understand its distribution, frequency, and characteristics.

This analysis provides a foundational understanding of customer demographics, vehicle attributes, and claim-related information before exploring relationships between multiple variables.

### Analysis 1: Claim Status Distribution

```python
# Claim Status Distribution

claim_counts = dataset['Claim Status Label'].value_counts()
claim_counts
plt.figure(figsize=(6,4))
plt.bar(
    claim_counts.index,
    claim_counts.values,
    color=['#2E86DE', '#E74C3C']
)
plt.title("Claim Status Distribution")
plt.xlabel("Claim Status")
plt.ylabel("Number of Policies")
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.savefig(
    'Python Analysis Images/claim_status_distribution.png',
    dpi=300,
    bbox_inches='tight'
)
plt.show()
```

### Business Observation

The claim status distribution indicates the proportion of policies with and without insurance claims.

A significantly larger number of policies fall under the "No Claim" category, while only a smaller proportion result in claims. This is a typical pattern in the insurance industry and suggests that the company's customer base is predominantly composed of lower-risk policyholders.

Understanding this distribution establishes the foundation for identifying the characteristics associated with customers who are more likely to file claims.

# 12. Bivariate Analysis

Bivariate analysis examines the relationship between two variables to identify patterns and factors that influence insurance claims.

The objective is to understand how customer demographics, vehicle characteristics, and regional factors are associated with claim occurrence. These insights help insurance companies identify high-risk segments and support data-driven decision-making.

### Analysis 1: Customer Age Group vs Claim Status

```python
age_claim = pd.crosstab(
    dataset['Customer Age Group'],
    dataset['Claim Status Label']
)
print(age_claim)

age_claim.plot(
    kind='bar',
    figsize=(8,5)
)
plt.title("Customer Age Group vs Claim Status")
plt.xlabel("Customer Age Group")
plt.ylabel("Number of Policies")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.savefig(
    'Python Analysis Images/Customer_Age_Group_VS_Claim_Status.png',
    dpi=300,
    bbox_inches='tight'
)
plt.show()
```

### Business Observation

The comparison highlights how insurance claims vary across different customer age groups. Identifying age segments with relatively higher claim frequencies enables insurers to design targeted risk assessment and pricing strategies.

### Analysis 2: Vehicle Age Group vs Claim Status

```python
vehicle_claim = pd.crosstab(
    dataset['Vehicle Age Group'],
    dataset['Claim Status Label']
)
print(vehicle_claim)
vehicle_claim.plot(
    kind='bar',
    figsize=(8,5)
)
plt.title("Vehicle Age Group vs Claim Status")
plt.xlabel("Vehicle Age Group")
plt.ylabel("Number of Policies")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.savefig(
    'Python Analysis Images/Vehicle_Age_Group_VS_Claim_Status.png',
    dpi=300,
    bbox_inches='tight'
)
plt.show()
```

### Business Observation

This analysis helps determine whether claim occurrence differs across vehicle age groups. If older vehicles exhibit higher claim frequencies, the company may consider additional risk evaluation during underwriting.

### Analysis 3: Fuel Type vs Claim Status

```python
# Fuel Type vs Claim Status

fuel_claim = pd.crosstab(
    dataset['fuel_type'],
    dataset['Claim Status Label']
)
print(fuel_claim)
fuel_claim.plot(
    kind='bar',
    figsize=(8,5),
    color=['#3498DB', '#E74C3C']
)
plt.title('Fuel Type vs Claim Status', fontsize=14, fontweight='bold')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Policies')
plt.xticks(rotation=0)
plt.legend(title='Claim Status')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

### Business Observation

This analysis compares insurance claim occurrence across different fuel types. Understanding these differences helps identify whether certain vehicle categories require additional risk assessment during underwriting.

### Analysis 4: NCAP Rating vs Claim Status

```python
# NCAP Rating vs Claim Status
ncap_claim = pd.crosstab(
    dataset['ncap_rating'],
    dataset['Claim Status Label']
)
print(ncap_claim)
ncap_claim.plot(
    kind='bar',
    figsize=(8,5),
    color=['#2ECC71', '#E74C3C']
)
plt.title('NCAP Rating vs Claim Status', fontsize=14, fontweight='bold')
plt.xlabel('NCAP Rating')
plt.ylabel('Number of Policies')
plt.legend(title='Claim Status')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

### Business Observation

Comparing claim occurrence across NCAP ratings helps evaluate whether vehicles with higher safety ratings experience fewer insurance claims. This insight supports risk-based pricing and underwriting decisions.

### Analysis 5: Region Density Group vs Claim Status

```python
# Region Density Group vs Claim Status

region_claim = pd.crosstab(
    dataset['Region Density Group'],
    dataset['Claim Status Label']
)
print(region_claim)
region_claim.plot(
    kind='bar',
    figsize=(9,5),
    color=['#9B59B6', '#E74C3C']
)
plt.title('Region Density Group vs Claim Status', fontsize=14, fontweight='bold')
plt.xlabel('Region Density Group')
plt.ylabel('Number of Policies')
plt.xticks(rotation=20)
plt.legend(title='Claim Status')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

### Business Observation

This analysis identifies how insurance claims differ across regions with varying population densities. Such insights help insurers understand geographical risk patterns and optimize regional pricing strategies.

### Analysis 6: Safety Feature Score vs Claim Status

```python
# Safety Feature Score vs Claim Status
plt.figure(figsize=(8,5))
dataset.boxplot(
    column='Safety Feature Score',
    by='Claim Status Label',
    grid=False
)
plt.title('Safety Feature Score vs Claim Status', fontsize=14, fontweight='bold')
plt.suptitle('')
plt.xlabel('Claim Status')
plt.ylabel('Safety Feature Score')
plt.show()
```

### Business Observation

The box plot compares the distribution of Safety Feature Scores between policies with and without claims. This helps assess whether vehicles equipped with stronger safety features demonstrate different claim patterns, providing valuable insights for underwriting and product design.

# 13. Multivariate Analysis

Multivariate analysis examines relationships among multiple numerical variables simultaneously. This helps identify potential correlations between customer characteristics, vehicle specifications, and safety-related features that may influence insurance claims.

```python
# Select numerical columns
numerical_df = dataset.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr_matrix = numerical_df.corr()

# Plot heatmap using Matplotlib
plt.figure(figsize=(14,10))

plt.imshow(corr_matrix, cmap='coolwarm', aspect='auto')

plt.colorbar(label='Correlation')

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)

plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

plt.title("Correlation Heatmap of Numerical Features")

plt.tight_layout()
plt.savefig(
    'Python Analysis Images/Correlation_HeatMap.png',
    dpi=300,
    bbox_inches='tight'
)
plt.show()
```

### Business Observation

The correlation heatmap indicates the strength of relationships between numerical variables within the insurance dataset.

Most variables exhibit weak to moderate correlations, suggesting that insurance claim behavior is influenced by multiple independent factors rather than a single dominant variable.
